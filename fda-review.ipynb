```r
# Train Logistic Regression model
log_model <- glm(y_train ~ ., data = as.data.frame(X_train), family = binomial)

# Predict probabilities on the test set
y_pred_prob <- predict(log_model, newdata = as.data.frame(X_test), type = "response")

# Convert probabilities to class predictions
y_pred <- ifelse(y_pred_prob > 0.5, 1, 0)

# Evaluate the model
confusionMatrix(factor(y_pred), factor(y_test))
```

```r
# Train Random Forest model
rf_model <- randomForest(X_train, as.factor(y_train), ntree = 100)

# Predict probabilities on the test set
y_pred_prob_rf <- predict(rf_model, X_test, type = "prob")[,2]

# Convert probabilities to class predictions
y_pred_rf <- ifelse(y_pred_prob_rf > 0.5, 1, 0)

# Evaluate the model
confusionMatrix(factor(y_pred_rf), factor(y_test))
```

```r
get_metrics <- function(pred, actual) {
  cm <- confusionMatrix(factor(pred), factor(actual))
  metrics <- list(
    Accuracy = cm$overall['Accuracy'],
    Precision = cm$byClass['Pos Pred Value'],
    Recall = cm$byClass['Sensitivity'],
    F1_Score = 2 * ((cm$byClass['Pos Pred Value'] * cm$byClass['Sensitivity']) / 
                      (cm$byClass['Pos Pred Value'] + cm$byClass['Sensitivity']))
  )
  return(metrics)
}
```

```r
# Store metrics for each model
# Convert predicted probabilities to binary class labels (0 or 1)
y_pred_log <- ifelse(y_pred_prob > 0.5, 1, 0)

log_metrics <- get_metrics(y_pred_log, y_test)  # Use class labels, not probabilities
tree_metrics <- get_metrics(y_pred_tree, y_test)
rf_metrics <- get_metrics(y_pred_rf, y_test)
knn_metrics <- get_metrics(y_pred_knn, y_test)
#svm_metrics <- get_metrics(y_pred_svm, y_test)
```

```r
# Normalize the training and test sets (min-max normalization)
normalize <- function(x) {
  return((x - min(x)) / (max(x) - min(x)))
}
```

```r
# Separate numeric and factor columns
numeric_features <- sapply(X_train, is.numeric)
categorical_features <- sapply(X_train, is.factor)
```

```r
# Normalize numeric features
X_train_norm <- as.data.frame(X_train)  # Ensure X_train is a data frame
X_test_norm <- as.data.frame(X_test)    # Ensure X_test is a data frame

X_train_norm[numeric_features] <- lapply(X_train[numeric_features], normalize)
X_test_norm[numeric_features] <- lapply(X_test[numeric_features], normalize)
```

```r
# Convert categorical features to dummy variables (one-hot encoding)
X_train_norm <- cbind(X_train_norm[numeric_features], model.matrix(~ . - 1, data = X_train[categorical_features]))
X_test_norm <- cbind(X_test_norm[numeric_features], model.matrix(~ . - 1, data = X_test[categorical_features]))
```

```r
# Train a neural network model
set.seed(123)
nn_model <- nnet(y_train ~ ., data = X_train_norm, size = 10, maxit = 200, linout = FALSE, trace = FALSE)

# Predict class probabilities on the test set
y_pred_prob_nn <- predict(nn_model, X_test_norm, type = "raw")

# Convert probabilities to class labels (0 or 1)
y_pred_nn <- ifelse(y_pred_prob_nn > 0.5, 1, 0)

# Evaluate the Neural Network model
nn_metrics <- get_metrics(y_pred_nn, y_test)
```
